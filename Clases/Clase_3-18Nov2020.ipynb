# Operaciones arítmeticas con imágenes

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread("D:/WILLAMS/Descargas/787930.jpg")
img2 = cv.imread("D:/WILLAMS/Descargas/Cartel Ventas/vera-davidova-I4WDjqFgHbE-unsplash.jpg")

#cambio de espacio de color
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)

#igualar el tamaño
fil,cols,chan = img1.shape
img2 = cv.resize(img2,(cols,fil))
```

```python
#Mostramos las imagenes

plt.figure(1)
plt.imshow(img1)
plt.title("Games")

plt.figure(2)
plt.imshow(img2)
plt.title("Mascarilla")
```

```python
#Combinar imagenes

img_out = cv.addWeighted(img1, 0.5, img2, 0.3, 0)
plt.imshow(img_out)
```
